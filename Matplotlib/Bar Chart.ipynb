```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company = ['Google','Amazon','IBM','Facebook']
revenue = [90,136,89,27]
ypos = np.arange(len(company))
ypos
plt.bar(ypos,revenue)
```

```python
company = ['Google','Amazon','IBM','Facebook']
revenue = [90,136,89,27]
ypos = np.arange(len(company))
ypos
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python
company = ['Google','Amazon','IBM','Facebook']
revenue = [90,136,89,27]
ypos = np.arange(len(company))
ypos
plt.xticks(ypos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos,revenue,label = 'Revenue')
plt.legend()
```

```python
profit = [40,2,34,12]
plt.xticks(ypos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos,revenue,label = 'Revenue')
plt.bar(ypos,profit,label = 'Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
plt.xticks(ypos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,label = 'Revenue')
plt.bar(xpos+0.2,profit,label = 'Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
plt.xticks(ypos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width = 0.4,label = 'Revenue')
plt.bar(xpos+0.2,profit,width = 0.4,label = 'Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
plt.yticks(ypos,company)

plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,height = 0.4,label = 'Revenue')
plt.barh(xpos+0.2,profit,height = 0.4,label = 'Profit')
plt.legend()
```
