```python
import matplotlib.pyplot as plt

exp_vals = [1400,600,300,410,250]
exp_labels = ['Home rent','Food','Phone/internet Bill','Car','Other utilities']
plt.pie(exp_vals , labels = exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels,radius=2)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels,radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels,radius=1.5,autopct='%0.2f%%',shadow = True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels,radius=1.5,autopct='%0.2f%%',shadow = True,explode = [0,0.2,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels = exp_labels,radius=1.5,autopct='%0.2f%%',shadow = True,explode = [0,0.2,0,0,0],startangle = 45)
plt.show()
```
