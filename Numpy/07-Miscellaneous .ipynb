### Boolean masking and advanced indexing 

```python
data = np.genfromtxt(r"E:\study materials\python\Data.txt",delimiter=',',dtype='int16')
data
```

#### To check if the data is greater than 50

```python
x = data > 50
x
```

#### To print the actual data

```python
v = data[data > 50]
print(v)
```

#### To check if any the values in the column is greater than 50

```python
x = np.any(data > 50,axis=0)
print(x)
```

#### To check if all the values in the column is greater than 50

```python
y = np.all(data > 50 ,axis = 0)
print(y)
```

#### To check if the value is greater than 50 and less than 100

```python
c = ((data > 50) & (data < 100))
c
```

#### To check id the value is not greater than 50 and less than 100

```python
d = (~((data > 50) & (data < 100)))
print(d)
```

Here ~ reperesents not
